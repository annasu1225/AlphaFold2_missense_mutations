Import packages and file

```python
import os.path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# import and label
root='/content/drive/MyDrive/IW Final/Final_Result'
file=os.path.join(root, 'Results_v3.csv')
```

Create a dataframe

```python
df=pd.read_csv(file)
df.head()
df
```

```python
W_id = df['wildtype_id'].tolist()
M_id = df['mutant_id'].tolist()
RMSD = df['RMSD'].to_numpy()
M_lDDT = df['mutant_lDDT'].to_numpy()
W_lDDT = df['wildtype_lDDT'].to_numpy()
N_lDDT = df['normalized_lDDT'].to_numpy()
Type = df['Type'].to_numpy()
```

```python
# number of type
num_monomer=(Type == 0).sum() 
num_multi=(Type == 1).sum() + (Type == 2).sum()
print(num_monomer)
print(num_multi)
```

Store data in arrays

```python
# get lddt and rmsd
M_mono_lddt = []
W_mono_lddt = []
mono_rmsd = []
M_multi_lddt = []
W_multi_lddt = []
multi_rmsd = []
N_mono_lddt = []
N_multi_lddt = []
for i in range(0,len(Type)):
  if Type[i]==0:
    M_mono_lddt.append(M_lDDT[i])
    W_mono_lddt.append(W_lDDT[i])
    N_mono_lddt.append(N_lDDT[i])
    mono_rmsd.append(RMSD[i])
  else:
    M_multi_lddt.append(M_lDDT[i])
    W_multi_lddt.append(W_lDDT[i])
    N_multi_lddt.append(N_lDDT[i])
    multi_rmsd.append(RMSD[i])

print('Mutant monomer lDDT: ' + str(M_mono_lddt))
print('Wildtype monomer lDDT: ' + str(W_mono_lddt))
print('Normalized monomer lDDT: ' + str(N_mono_lddt))
print('Monomer rmsd: ' + str(mono_rmsd))
print('Mutant multimer lDDT: ' + str(M_multi_lddt))
print('Wildtype multimer lDDT: ' + str(W_multi_lddt))
print('Multimer rmsd: ' + str(multi_rmsd))
print('Normalized multimer lDDT: ' + str(N_multi_lddt))
```

lDDT Distribution

```python
sns.set()

# Set up a figure with set of axes
fig, ax = plt.subplots(1, 1)

# Add axis labels
ax.set_xlabel('lDDT')
ax.set_ylabel('count');

# Make bin boundaries
bins = np.arange(0.6, 1.0, 0.05)

# Make histogram
_ = ax.hist((mono_lddt, multi_lddt), bins=bins)

# Add a legend
plt.legend(('Monomers', 'Multimers'), loc='upper left');

plt.savefig('histogram.png')
```

```python
sns.set()

# Set up a figure with set of axes
fig, ax = plt.subplots(1, 1)

# Add axis labels
ax.set_xlabel('Mutant lDDT')
ax.set_ylabel('count');

# Make bin boundaries
bins = np.arange(0.6, 1.0, 0.05)

# Make histogram
_ = ax.hist((M_mono_lddt, M_multi_lddt), bins=bins, density=True)

# Add a legend
plt.legend(('Monomers', 'Multimers'), loc='upper left');

plt.savefig('histigram_normalized.png', dpi=1000)
```

```python
mono_counts, bin_edges = np.histogram(mono_lddt, bins=8)
print(mono_counts)
mono_p = mono_counts / num_monomer
print(mono_p)

multi_counts, bin_edges = np.histogram(multi_lddt, bins=8)
print(multi_counts)
multi_p = multi_counts / num_multi
print(multi_p)
```

Correlation plot for Mutant lDDT vs RMSD

```python
corr = sns.scatterplot(data=df, x="RMSD", y="mutant_lDDT", hue="Type", palette="deep")
handles, labels  =  corr.get_legend_handles_labels()
corr.legend(handles, ['Monomer', 'Homomer', 'Heteromer'], loc='lower right')
plt.ylabel('Mutant lDDT');
plt.savefig('RMSD_correlation.png', dpi=1000)
```

Correlation plot for Mutant lDDT vs Wildtype lDDT

```python
type_corr = sns.scatterplot(data=df, x="wildtype_lDDT", y="mutant_lDDT", hue="Type", palette="deep")
type_handles, labels  =  type_corr.get_legend_handles_labels()
type_corr.legend(type_handles, ['Monomer', 'Homomer', 'Heteromer'], loc='lower right')
plt.ylabel('Mutant lDDT');
plt.xlabel('Wildtype lDDT');
plt.savefig('Wildtype_correlation.png', dpi=1000)
```

Normalized lDDT plot

```python
sns.set()

# Set up a figure with set of axes
fig, ax = plt.subplots(1, 1)

# Add axis labels
ax.set_xlabel('Mutant lDDT / Wildtype lDDT')
ax.set_ylabel('count');

# Make bin boundaries
bins = np.arange(0.8, 1.2, 0.05)

# Make histogram
_ = ax.hist((N_mono_lddt, N_multi_lddt), bins=bins, density=True)

# Add a legend
plt.legend(('Monomers', 'Multimers'), loc='upper left');

plt.savefig('normalized_histigram_normalized.png')
```

```python
fig, ax = plt.subplots()

N_mono_counts, N_mono_bins = np.histogram(N_mono_lddt, 10, density=True)
N_multi_counts, N_multi_bins = np.histogram(N_multi_lddt, 10, density=True)

ax.plot(N_mono_bins[:-1], N_mono_counts)
ax.plot(N_multi_bins[:-1], N_multi_counts)

# Add axis labels
ax.set_xlabel('Mutant lDDT / Wildtype lDDT')
ax.set_ylabel('count');

# Add a legend
plt.legend(('Monomers', 'Multimers'), loc='upper left');

plt.savefig('N_lddt.png', dpi=1000)
```
